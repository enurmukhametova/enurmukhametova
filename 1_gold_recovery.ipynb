# Технологический процесс

Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.


**Стадии процесса:**
- **Флотация**

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

- **Очистка**

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.


### Описание данных

**Технологический процесс**
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
   - Xanthate **— ксантогенат (промотер, или активатор флотации);
   - Sulphate — сульфат (на данном производстве сульфид натрия);
   - Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

**Параметры этапов**
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

**Наименование признаков**

Наименование признаков должно быть такое:
[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag

Возможные значения для блока [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

Возможные значения для блока [тип_параметра]:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### 1.2 Проверка эффективности обогащения

Необходимо вычислить эффективность обогащения на обучающей выборке для признака rougher.output.recovery. 

```python
# доля золота в концентрате после флотации/очистки
c = df_train['rougher.output.concentrate_au']

# доля золота в концентрате/сырье до флотации/очистки
f = df_train['rougher.input.feed_au']

# доля золота в отвальных хвостах флотации/очистки
t = df_train['rougher.output.tail_au']
```

```python
# создадим новый столбец recovery_calculation в обучающей выборке для вычислений
df_train['recovery_calculation'] = (c * (f - t)) / (f * (c - t)) * 100

# посчитаем среднее по столбцу recovery_calculation
df_train['recovery_calculation'].mean()
```

```python
# посчитаем среднее по столбцу rougher.output.recovery (целевой признак)
df_train['rougher.output.recovery'].mean()
```

```python
if round(df_train['recovery_calculation'].mean(), 5) == round(df_train['rougher.output.recovery'].mean(), 5):
    print('Значения совпадают, эффективность расчитана правильно.')
else:
    print('Значения не совпадают, эффективность расчитана неверно.')
```

Найдем MAE между нашими расчетами и значением целевого признака.

```python
print('Значение MAE между нашими расчетами и значением целевого признака:', 
      mean_absolute_error(df_train['rougher.output.recovery'], df_train['recovery_calculation']))
```

Значение MAE очень маленькое (возможно, дело в том, как хранится информация), поэтому делаем вывод, что эффективность расчитана верно.

### 1.4  Предобработка данных

Как мы уже выяснили ранее, датафреймы содержат пропущенные значения. Данные индексируются датой и временем получения информации (признак date). Соседние по времени параметры часто похожи, поэтому заполним пропуски при помощи KNNImputer. 

Класс KNNImputer обеспечивает восстановление пропущенных значений с использованием подхода к-ближайших соседей. Каждый отсутствующий объект рассчитывается с использованием значений от n_neighbors ближайших соседей, у которых есть значение для данного объекта. Характеристики соседей усредняются равномерно или взвешиваются по расстоянию до каждого соседа. То есть выбирается k-точек, которые больше всего похожи на рассматриваемую, и уже на их основании выбирается значение для пустой ячейки.

```python
# проверим количество пропусков в целевых признаках обучающей выборки
print('Количество пропусков в целевых признаках обучающей выборки:\n', 
      df_train[TARGETS].isna().sum())
```

```python
# определяем импортер c количеством соседних образцов, используемых для интерполяции, равным 5
# и весовой функцией, используемой при прогнозировании, равной uniform (однородные веса)
knn_imputer = KNNImputer(n_neighbors=5, weights = 'uniform')

# устанавливаем импортер на df_train (без столбца date) при помощи fit 
# восстанавливаем данные при помощи transform
# полученные данные преобразовываем в датафрейм
df_train = pd.DataFrame(knn_imputer.fit_transform(df_train.drop(['date'], axis = 1)),
             columns = df_train.drop(['date'],axis = 1).columns,
             index = df_train.index)

print('Количество пропусков в df_train:', 
      df_train.isnull().sum().sum())
```

## 3 Модель

### 3.1 Вычисление итоговой sMAPE

Напишем функцию для вычисления итоговой sMAPE. Воспользуемся следующими формулами:

```python
def smape(target, predicted):
    return 1/len(target) * np.sum(2 * np.abs(target - predicted) / (np.abs(target) + np.abs(predicted))*100)
def smape_final(rougher, final):
    return 0.25 * rougher + 0.75 * final
```

**Объявим скорер для кросс-валидации.**

```python
smape_scorer = make_scorer(smape, greater_is_better=False)
```

Вычислим sMAPE для этапа флотации.

Вычислим sMAPE для финального этапа.

```python
model_dummy_f = DummyRegressor(strategy = 'median')
model_dummy_f.fit(df_train_features, df_train_target_f)

predict_dummy_model_f = cross_val_score(model_dummy_f,
                                        df_train_features,
                                        df_train_target_f, 
                                        cv = CV, 
                                        scoring = smape_scorer).mean()
predict_dummy_model_f.round(2)
```

**Вычислим итоговую sMAPE для простейшей модели dummy.**

```python
dummy_model_smape = smape_final(predict_dummy_model_r, predict_dummy_model_f)
print('Итоговое значение sMAPE простейшей модели: ', -1 * dummy_model_smape)
```

Напишем функцию проверки качества модели методом кросс-валидации.

```python
def cross_validation(model):
    '''Функция проверки качества модели методом кросс-валидации'''
    score_val_r = cross_val_score(model, 
                                 df_train_features, 
                                 df_train_target_r, 
                                 cv = CV, 
                                 scoring = smape_scorer).mean() 
    
    print('Cреднее значение sMAPE для этапа флотации:', -1 * score_val_r.round(2))
    
    score_val_f = cross_val_score(model, 
                                   df_train_features, 
                                   df_train_target_f, 
                                   cv = CV, 
                                   scoring = smape_scorer).mean() 
    
    print('Cреднее значение sMAPE для финального этапа: ',-1*score_val_f.round(2))
    
    final_score_smape = smape_final(score_val_r, score_val_f)
    
    print('Итоговое sMAPE модели на кросс-валидации: ', -1*final_score_smape.round(2)) 
```

Обучим модели и проверим, какая из них даст минимальную sMAPE.
